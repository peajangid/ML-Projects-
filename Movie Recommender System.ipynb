```python
movies = movies[['genres','id','keywords','title','overview','cast','crew']]
```

### Data preprocessing

```python
movies.isnull().sum()
```

```python
## Overview columns has 3 missing values so we can drop them
movies.dropna(inplace=True)
```

```python
## To check for duplicated values
movies.duplicated().sum()
movies.shape
```

```python
import ast
## Genres
movies.genres[1]

## The given format is a list of dictionaries
## So tocreating a helper function
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
## similarly for keywords column
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
           L.append(i['name'])
           counter +=1
        else:
           break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
## Fetching the director name from the cast column
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
    return L
            
            
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'][0]
```

```python
## Now creating a list of all the words in overview 
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
# Now here every coll is in list so merging these lists 
# But before that removing any space in 2 words of a single data value 
# so as to not create 2 tags for just one man
# 'Sam Wordington' to 'samwordington'
movies
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(' ','') for i in x])
movies['keywords'] =movies['keywords'].apply(lambda x: [i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
# creating a new col that is tags which is a concatenation of overview, genres, keywords, cast and crew
movies['tags'] = movies['overview']+movies['genres']+movies['cast']+movies['crew']+ movies['keywords']
movies.head()
```

```python
#dropping the other cols 
new_df = movies[['id','title','tags']]
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: ' '.join(x))
```

```python
# now convert the tags in lower
new_df['tags'] = new_df['tags'].apply(lambda x :x.lower() ) 
new_df['tags']
```

```python
# vectorizing 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
# Now we apply stemming using nltk for stemming
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
    ps= PorterStemmer()
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
    
```

```python
new_df['tags'].apply(stem)
```

```python
# NOw we calcuate the angular distance but instead of euclidean distance because in higher dim euclidean distance is not a reliable measume
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
## similarity is an array of matrics that gives the similarity score
similarity

# now to get similar 5 movies 
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x: x[1])
# 1. here we sorted to get the most similar movies 
# 2. we used enumerate function so we can get the movie index as well to get the movie index of the similar movies
# 3. also we used key because the sorting was 1st on basis of index which we changed to the similarity score
```

```python

def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
